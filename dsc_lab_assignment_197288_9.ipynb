# PCA FROM SCRATCH

```python
import numpy as np
import pandas as pd
from sklearn.datasets import load_iris
```

```python
data=load_iris()
X=data.data
y=data.target
```

```python
X
```

```python
# scaling the dataset
def standardise(X):
    mu=np.mean(X,axis=0)
    std=np.std(X,axis=0)
    X=(X-mu)/std
    return X
X=standardise(X)
```

```python
# calculating covariance matrix
cov=np.cov(X.T)
cov.shape
```

```python
# calculating eigen values and eigen vectors
eig_val,eig_vec=np.linalg.eig(cov)
sorted_idxs=np.argsort(eig_val)
eig_val=np.sort(eig_val)
j=1
for i in range(len(eig_val)-1,-1,-1) :
    print("percentage of explained variance for principle component {} = {}".format(j,eig_val[i]/np.sum(eig_val)))
    j+=1;
```

```python
# transforming data 
for k in range(1,len(eig_val)):
    new_data=np.matmul(X,eig_vec[sorted_idxs[:k]].T)
    print(new_data)
```

```python
from sklearn.linear_model import LogisticRegression as LogReg
import seaborn as sb
```

```python
data=pd.read_csv("framingham.csv")
data=data.dropna()
data.head()
```

```python
X=data.iloc[:,:-1].values
y=data.iloc[:,-1].values
X=standardise(X)                      # scaling data set
cov=np.cov(X.T)                       # calculating covariance matrix
eig_val,eig_vec=np.linalg.eig(cov)    # getting eigen values and eigen vectors
sorted_idxs=np.argsort(eig_val)       # getting sorted indices of eigen values
eig_val=np.sort(eig_val)
sum=0
j=0;
for i in range(len(eig_val)-1,-1,-1) :
    sum+=(eig_val[i]/np.sum(eig_val))
    j+=1 
    if(sum>0.98): break                          # if current variance information exceeded 98% percentage we can stop
X_new=np.matmul(X,eig_vec[sorted_idxs[:j]].T)    # transformed data
X_new.shape
```

```python
model=LogReg()
model.fit(X_new,y)
model.score(X_new,y)
```

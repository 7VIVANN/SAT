```python
import numpy as np
import pandas as pd
import random
import matplotlib.pyplot as plt
```

```python
def SplitDataSet(X,y,p=0.8):  # function to split data set
    n,_=X.shape
    train_idxs=random.sample(range(0,n-1),int(np.ceil(p*n)))  # generating random numbers 
    train_idxs=np.sort(train_idxs)
    X_train=X[train_idxs]
    y_train=y[train_idxs]
    test_idxs=[i for i in range(n) if i not in train_idxs]
    X_test=X[test_idxs]
    y_test=y[test_idxs]
    return X_train,y_train,X_test,y_test
def standardise(X):
    mu=np.mean(X,axis=0)
    std=np.std(X);
    return (X-mu)/std
```

```python
class ANN:
    def __init__(self,m_input,m_output,hidden_sizes=[30]):
        self.mx=m_input
        self.my=m_output
        self.nh=len(hidden_sizes)
        self.sizes=[self.mx]+hidden_sizes+[self.my]
        self.W={}
        self.B={}
        self.cost=[]
        for i in range(self.nh+1):
            self.W[i+1]=np.zeros((self.sizes[i],self.sizes[i+1]))
            self.B[i+1]=np.zeros((1,self.sizes[i+1]))
    def relu(self,x):
        return np.maximum(0,x)
    def grad_relu(self,x):
        return np.where((x>0),1,0)
    def predict(self,X):
        y_pred=[]
        for x in X:
            self.forward_prop(x)
            y_pred.append(self.H[self.nh+1]);
        return y_pred
    def cal_loss(self,X,y):
        y_pred=self.predict(X)
        return np.mean((y_pred-y)**2)
    def plotCost(self):
        fig=plt.figure(figsize=(5,5))
        epochs_arr=np.arange(len(self.cost))
        plt.plot(epochs_arr,self.cost)
        plt.xlabel("Number of Epochs")
        plt.ylabel("RMS error")
        plt.title("RMS error vs Number of Iterations")   
    def forward_prop(self,X):
        self.H={}
        self.H[0]=X.reshape(1,-1);
        for i in range(1,self.nh+2):
            self.H[i]=self.relu(np.matmul(self.H[i-1],self.W[i])+self.B[i])
    def backprop(self,y):
        self.dW={}
        self.dB={}
        self.dH={}
        self.dA={}
        L=self.nh+1
        self.dA[L]=self.H[L]-y
        for i in range(L,0,-1):
            self.dW[i]=np.matmul(self.H[i-1].T,self.dA[i])
            self.dB[i]=self.dA[i]
            self.dH[i-1]=np.matmul(self.dA[i],self.W[i].T)
            self.dA[i-1]=np.multiply(self.dH[i-1],self.grad_relu(self.H[i-1]))
    def fit(self,X,Y,alpha=0.01,num_iters=500):
        for e in range(num_iters):
            dW = {}
            dB = {}
            m=X.shape[1]
            for i in range(self.nh+1):
                dW[i+1] = np.zeros((self.sizes[i], self.sizes[i+1]))
                dB[i+1] = np.zeros((1, self.sizes[i+1]))
            for x, y in zip(X, Y):
                self.forward_prop(x)
                self.backprop(y)
                for i in range(self.nh+1):
                    dW[i+1] += self.dW[i+1]
                    dB[i+1] += self.dB[i+1]
            for i in range(1,self.nh+2):
                self.W[i]-=alpha*(dW[i]/m)
                self.B[i]-=alpha*(dB[i]/m)
            self.cost.append(self.cal_loss(X,y))
        
```

# BOSTON DATASET

```python
data=pd.read_csv("boston.csv")
data
```

```python
X=data.iloc[:,:-1].to_numpy()
X=standardise(X)
y=data.iloc[:,-1].to_numpy()
y=y.reshape((y.shape[0],1))
X_train,y_train,X_test,y_test=SplitDataSet(X,y)

ann=ANN(X_train.shape[1],y_train.shape[1])
ann.fit(X_train,y_train)
ann.plotCost()
print("Cost for training data = {}".format(ann.cost[-1]))
print("Cost for testing data = {}".format(ann.cal_loss(X_test,y_test)))
```
